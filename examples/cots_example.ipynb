# Commercial Off The Shelf (COTS) Car
The `COTSCar` class allows you to configure a car based only on discrete integer selections. THis means that the complete design is represented as a vector with 17 elements:

1.   Rear Wing Material (with assumed/fixed dimensions)
2.   Front Wing Material (with assumed/fixed dimensions)
3.   Side Wing Material (with assumed/fixed dimensions)
4.   Cabin Material (with assumed/fixed dimensions)
5.   Impact Attenuator Material (with assumed/fixed dimensions)
6.   Rear Tire Selection
7.   Front Tire Selection
8.   Engine Selection
9.   Brake Selection
10.  Suspension Selection
11.  Impact attenuator
12.  Cabin
13.  Pressure rating of front tire
14.  Pressure rating of rear tire
15.  Rear wing
16.  Front wing
17.  Side wings



The available materials are:

```
sae.materials
```

The available tires are:

```
sae.tires
```

The available engines are:

```
sae.motors
```

The available brakes are:

```
sae.brakes
```

The available subspension options are:

```
sae.suspension
```

The available impact attenuators are:

```
sae.attenuators
```

The available cabins are:

```
sae.cabins
```

The available tire pressure ratings are:

```
sae.pressure
```

And the available wings are:

```
sae.wings
```

You can generate a randomly-configured `COTSCar` and evaluate it like this:

```
car = sae.COTSCar()
print(car.vector)
print(car.objectives())
print(car.constraints_bound())
print(car.constraints_lin_ineq())
print(car.constraints_nonlin_ineq())
```

You can directly reset the vector that defines the car like this:

```
car.set_vec([1]*17)
car.objectives()
```
